## Operaciones Vectoriales

### Suma de vectores

La suma de vectores cumple con la siguientes propiedades:  
- $\vec{A}+\vec{B}=\vec{B}+\vec{A}$  
- $\left(\vec{A}+\vec{B}\right)+\vec{C}=\vec{A}+\left(\vec{B}+\vec{A}\right)$  
- Con $c\in \mathbb{R}$, $c\left(\vec{A}+\vec{B}\right)=c\vec{A}+c\vec{B}$  
**Comprobemos**:   
- Todas las propiedades de la suma  
- ¿Que pasa si $c<0$?  

### Componentes de un vector  
Las componentes de un vector son vectores rectangulares que permiten realizar el proceso de suma de manera más sencilla y generalizada.  
<p align="center">
  <img src="Figuras/Componentes.png" alt="Vector" width="300" height="200">
</p>  
De manera que cualquier vector puede ser escrito en sus componentes rectangulares,  
\begin{equation}
\vec{A}=\vec{A_{x}}+\vec{A_{y}}
\end{equation}  
y la suma de 2 vectores se puede establecer, en el contexto de sus componentes, de la siguiente forma:  
\begin{equation}
\vec{A}+\vec{B}=\vec{A_{x}}+\vec{A_{y}}+\vec{B_{x}}+\vec{B_{y}}
\end{equation}  

### Vector unitario  

Un vector unitario es aquel que sirve para designar el sentido y dirección que tiene magnitud $1$. Por ejemplo, para designar la dirección a lo largo de los ejes $x$ y $y$ se acostumbra a usar la notación $\hat{x}$ y $\hat{y}$. Otra alternativa consiste en usar las letras $\mathbf{i}$ y $\mathbf{j}$. Se debe tener cuidado en que una vez se utilice una de las notaciones seleccionadas no se debe mezclar con la otra. Teniendo en cuenta lo anterior, veamos lo siguiente:  
\begin{equation}
\vec{A}=(A_{x},A_{y})=A_{x}\hat{x}+A_{y}\hat{y}=A_{x}\mathbf{i}+A_{y}\mathbf{j}
\end{equation}

### Magnitud de un vector
Una vez se establecidas las componentes rectangulares de un vector, es sencillo establecer que la magnitud de cualquier vector a partir del **Teorema de Pitagoras** de la siguiente forma:
\begin{equation}
A=\sqrt{A_{x}^{2}+A_{y}^{2}}
\end{equation}
Para determinar la dirección del vector es necesario establecer respecto a que eje. Por ejemplo si tomamos la dirección respecto al eje $x$, la dirección se puede establecer de la siguiente forma
\begin{equation}
\theta=\arcsin\left(\frac{A_{y}}{A}\right)=\arccos\left(\frac{A_{x}}{A}\right)=\arctan\left(\frac{A_{y}}{A_{x}}\right)
\end{equation}

### Producto escalar de 2 vectores  
El **producto escalar** de 2 vectores se denota por $\vec{A}\cdot\vec{B}$ y se determina por medio de alguna de las siguientes expresiones:  
\begin{equation}
\vec{A}\cdot\vec{B}=AB\cos\theta=A_{x}B_{x}+A_{y}B_{y}
\end{equation}

### Producto vectorial de 2 vectores  
El **producto vectorial** de 2 vectores se denota por $\vec{A}\times\vec{B}$ y se determina por medio de alguna de las siguientes expresiones:  
\begin{equation}
\vec{A}\times\vec{B}=AB\sin\theta \hat{z}=\begin{vmatrix}
\hat{x}&\hat{y}\\
A_{x} & A_{y} \\ 
B_{x} & B_{y}
\end{vmatrix}
\end{equation} 
Para determinar el resultado en componentes se debe tern en cuenta las siguientes relaciones:
\begin{align}
\hat{x}\times\hat{y}=\hat{z}\\
\hat{z}\times\hat{x}=\hat{y}\\
\hat{y}\times\hat{z}=\hat{x}
\end{align}

### Propiedades de los productos vectoriales  
|$\vec{A}\cdot\vec{B}$ | $\vec{A}\times\vec{B}$ |
|---------------------------|-------------------------------|
|$\vec{A}\cdot\vec{B}=\vec{B}\cdot\vec{A}$   | $\vec{A}\times\vec{B}=\vec{A}\times\vec{B}$   |
|$c(\vec{A}\cdot\vec{B})=(c\vec{A})\cdot\vec{B}$   | $c(\vec{A}\times\vec{B})=(c\vec{A})\times\vec{B}$  |
|$\vec{A}\cdot(\vec{B}+\vec{C})=\vec{A}\cdot\vec{B}+\vec{A}\cdot\vec{C}$ | $\vec{A}\times(\vec{B}+\vec{C})=\vec{A}\times\vec{B}+\vec{A}\times\vec{C}$ |
